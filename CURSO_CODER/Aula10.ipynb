                                                 FUNÇÃO DE AGREGAÇÃO

Usada para resumir e analisar dados em uam serie ou dataframe.

```python
import pandas as pd
import numpy as np
```

```python
fastfood = pd.read_excel('fastfood.xlsx')
```

```python
fastfood.head(2)
```

```python
print('count:', fastfood['protein'].count())
print('maximo:', fastfood['protein'].max())
print('minimo:', fastfood['protein'].min())
print('mean:', fastfood['protein'].mean())
print('std:', fastfood['protein'].std())
print('var:', fastfood['protein'].var())
```

```python
fastfood.mean()
```

```python
fastfood.describe().round(1)
```

                                                    GROUP BY

Funcao de agregacao que permite agrupar os dados de um dataframe com bgase em uma ou mais colunas, e, em seguida, aplocar funcoes de agregacao a esses grupos.

```python
fastfood.groupby('restaurant')[['calories']].mean().round(2) #media de calorias por restaurantes
```

```python
fastfood.groupby('restaurant')[['calories','sugar']].describe().round(2)
```

GROUP BY - AGG

```python
fastfood.groupby('restaurant').agg({'calories': ['mean','max'], 'protein':['min']})
```

                                                        ORDENAÇÃO

Ordenar o DataFrame com base em uma ou mais colunas.

```python
fastfood.sort_values(['restaurant', 'calories']).head(4)
```

```python
fastfood.sort_values('calories', ascending = False).head(3) #ordenas do maior pro menor
```

```python
fastfood.sort_values(['restaurant', 'calories'], ascending= [True, False] ).head(4) #ordena o restaurant e as calories depois
```

```python
fastfood.sort_values(by = ['calories'], ignore_index = True).head(3) #mantem os indices de 0 em diante
```

                                                     merge (join)

Permite combinar dois ou mais DataFrames com base em uma ou mais colunas comuns, chamadas chaves.

```python
df3 = pd.DataFrame({'A': [1,2,3], 'B': [4,5,6]})

df4 = pd.DataFrame({'A': [1,2,3,4], 'C': [1,2,3,4]})
```

```python
df3
```

```python
df4
```

Fazer um merge pra trazer a coluna C do df4 para junto do df2.

```python
df3.merge(df4, on=['A'], how= 'left') #so puxou o C ate a linha 3 pois a df3 so tem 3 linhas.
```

```python
df4.merge(df3, on=['A'], how= 'left') #puxou da df3 pra df4, como df3 so tme 3 linhas, a ultima ficou NaN
```

```python
df4.merge(df3,left_on=['C'], right_on=['A'], how= 'left')
```
