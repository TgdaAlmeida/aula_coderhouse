COMPUTAR CADA ESTATISTICA DO DESCRIBE INDIVIDUALMENTE

```python
data_final['PREÇO MÍNIMO REVENDA'].min()
```

```python
mean = data_final['PREÇO MÍNIMO REVENDA'].mean()
std = data_final['PREÇO MÍNIMO REVENDA'].std()
print(f'A media dos preços minimos de revenda e {mean:.2f} e +- {std:.2f}')
```
