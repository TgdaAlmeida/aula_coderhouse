                                                       FILTROS, LOC E ILOC

FILTRO: Filtra dados com base em uma coleção booleana.

```python
df2[df2['calories'] < 100].head(3)
```

LOC: Usado para acessar linhas e colunas em um DataFrame usando rotulos. O metodo aceita dois parametros: o 1 especifica as linhas que devem ser acessadas e o 2 especifica as colunas.

```python
df2.loc[0:2, ['calories','item','restaurant']] #ESCOLHENDO INDICE DE 0 A 2 (INCLUI A 2) E A COLUNA ITEM
```

ILOC: Usado para acessar linhas e colunas usando indices inteiros. O 1 parametro especifica as linhas acessadas e o 2 especidica as colunas.

```python
df2.iloc[0:2,0:3] #SO TRABALHA COM INDICES
```

```python
df2.info()
```

```python
df2.isnull() #ONDE FOR NULO O BOOL RETORNA TRUE
```

```python
df2['calcium'].isnull().head() #ATRIBUINDO A VARIAVEL APARECE OS DADOS
```

DROPNA() Remove as linhas que tem valor nulo.

```python
df2.dropna() #ONDE TIVER VALOR NULO A LINHA SERA REMOVIDA, CUIDADO!
```

```python
df2.dropna(subset='protein').head(3) #SOMENTE NA COLUNA PROTEINA, SE TIVER NULO VAI REMOVER A LINHA
```
