                                                        MATPLOTLIB

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

1 - Grafico de linhas: Usado para demonstrar a evolução de um dado ao longo do tempo. ex. o preço de uma ação ao longo do tempo.

```python
x = [1,2,3,4,5]
y = [10,8,6,4,2]

plt.plot(x,y)

plt.show()
```

2 - Grafico de barras: Usado para comparar dadps em diferentes categorias. Usado com dados categoricos. ex. Um restaurante comparado ao numero de calorias.

```python
x = ['A','B','C','D','E']
y = [3,7,1,8,4]
plt.bar(x,y)
plt.show()
```

3 - Grafico de pizza: Mostra a distribuição percentual de dados em uma unica categoria.

```python
labels = ['Maças', 'Bananas', 'Laranjas', 'Morangos']
quantidades = [20, 25, 10, 30]
plt.pie(quantidades, labels=labels)
plt.show()
```

4 - Grafico de dispersao: mostra a relacao entre duas variaveis numericas. 2 variaveis numericas. ex. Calorias x proteina.

```python
x = [1, 2, 3, 4, 5]
y = [10, 8, 6, 4, 2]
plt.scatter(x,y)
plt.show()
```

5 - Histograma: mostra a distribuição de frenquencia de dados (numericos) em um intervalo.

```python
dados = [1,3,3,4,4,4,5,5,6,7,8,9]
plt.hist(dados, bins=4)
plt.show()
```

6 - Grafico de area: mostra a tendencia e a magnitude da mudança ao longo do tempo.

```python
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.fill_between(x, y)
plt.show()
```

                                                       CONFIGIRAÇÕES 

Título do gráfico: função "title()"     .    
Etiquetas dos eixos: funções "xlabel()" e "ylabel()"      . 
Limites dos eixos: funções "xlim()" e "ylim()"    .
Escala dos eixos: funções "xscale()" e "yscale()"    .
Legenda: função "legend()"    .
Cores: função "color" ou "c"    .

```python
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]


plt.plot(x, y,
         color ="b",
         linestyle ="--",
         linewidth = 2,
         marker= "o",
         markersize = 7,
         markerfacecolor = "w")
plt.title('Calories x Protein')
plt.xlabel('Calories')
plt.ylabel('Protein')
plt.xlim(0,6)
plt.ylim(0,12)
plt.grid(True)
```

```python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)

#criando os subplots

fig, axs = plt.subplots(3,1, figsize = (18,18))
axs[0].plot(x,y1)
axs[0].set_title('Grafico Seno')
axs[0].set_xlabel('Baixo')
axs[0].set_ylabel('Lateral')
axs[1].plot(x,y2)
axs[1].set_title('Grafico Cosseno')
axs[2].plot(x,y3)
axs[2].set_title('Grafico Tangente')
```

                                                                SEABORN

O Seaborn é uma biblioteca de visualização de dados baseada no Matplotlib. Enquanto o Matplotlib é uma biblioteca mais básica e flexível, o Seaborn oferece uma interface mais simples e especializada par a criação de graficos estatisticos atraentes e informativos.
