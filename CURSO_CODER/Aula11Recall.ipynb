```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

                                                       ORIENTADO A ESTADO

```python
fastfood = pd.read_excel('fastfood.xlsx')
```

```python
fastfood_total_pratos = fastfood.groupby('restaurant')['item'].count().reset_index() #quantos pratos tem em cada restaurante
fastfood_total_pratos
```

```python
plt.bar(fastfood_total_pratos['restaurant'], fastfood_total_pratos['item'], color = 'k')
plt.xticks(rotation = 45) #faz a rotacao dos titulos orientados a estado
plt.grid()
plt.show()
```

```python
plt.scatter(fastfood['calories'], fastfood['protein'], color = 'b')
plt.title('Titulo')
plt.xlabel('calorias')
plt.ylabel('proteinas')
plt.show()
```

```python
x = [2, 2, 3, 4, 5, 6, 7, 9, 10]
y = [2, 3, 4, 3, 4, 5, 6, 7, 8]

plt.plot(x, y, label = 'Desvio', linestyle = ':', color = 'm', lw = 2.0, markersize = 2 )
plt.title('Gráfico')
plt.xlabel('Elixo X')
plt.ylabel('Eixo Y')
plt.legend()
#plt.xticks([1, 5, 10, 15]) da exatamente os marcadores
#plt.yticks([1, 5, 10, 15])
plt.axis( xmin = 0, xmax = 12, ymin = 0, ymax = 12)
#plt.axis('auto') cria uma margem automatica

plt.show()
```
