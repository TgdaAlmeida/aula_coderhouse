SELECIONANDO REGISTROS DOS ANOS DE 2008 2010 E 2012

ALTERNATIVA 1

```python
data.head(2)
```

```python
data['DATA INICIAL']
```

```python
data['ANO'] = data['DATA INICIAL'].str.extract(r'(\d{4})')
```

```python
del data['Ano']
```

```python
data.head(2)
```

```python
data['ANO'].unique()
```

```python
data['ANO'] = data['ANO'].astype(int)
```

```python
#cria uma variavel alocando todas as comparações 
selecao = (data['ANO'] == 2008) | (data['ANO'] == 2010) | (data['ANO'] == 2012)
```

```python
selecao
```

```python
data[selecao]['ANO'].unique()
```

ALTERNATIVA 2

```python
lista_de_anos = [2008, 2010, 2012]
```

```python
#retorna uma series de booleanos
selecao = data['ANO'].isin(lista_de_anos)
selecao
```

```python
data[selecao]
```
