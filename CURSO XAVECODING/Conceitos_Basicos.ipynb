MÉTODO ILOC (Seleção baseada em indice)

```python
#selecionando a linha 1 do DF
data.iloc[1] #o 'i' e iloc significa index
```

Selecionando multiplas linhas atraves do fatiamento (slicing)

```python
#selecionando as linhas de indice de 0 ate 5 (incluso)
data.iloc[:6] # :6 configira de 0 ate 6, nao contando o 6
```

```python
#selecionando as linhas de indice de 10 a 15 (incluso)
data.iloc[10:16]
```

```python
#selecionando as linhas de indices 1, 5, 10, 15
data.iloc[[1, 5, 10, 15]] #passando uma lista de indices separadamente
```

```python
#selcionando as linhas de indices 5, 1, 15, 10
data.iloc[[5, 1, 15, 10]]
```

```python
#retornar o valor da linha de indice 1 e  coluna 3 ('ESTADO')
data.iloc[1, 3] #primeiro valor é o indice e segundo valor a coluna para fazer o cruzamento
```

MÉTODO LOC (Seleção baseada em rotulos) primeiro linha, depois coluna

```python
pesquisa
```

```python
pesquisa.iloc[0] #pega a linha do indice 0 (implicito) usando o iloc
```

```python
pesquisa.iloc[0, 1] #pega o valor da linha do indice 0 (implicito) e coluna 1 usando o iloc
```

```python
#retorna a linha cujo rotulo do index é 'XboxOne'
pesquisa.loc['XboxOne']
```

```python
#caso que nao funciona, iloc acessando com rotulos
pesquisa.iloc['XboxOne'] #nao funciona pois iloc usa indices numericos
```

```python
#caso que nao funciona, loc acessando com indices
pesquisa.loc[0] #nao funciona pois loc usa rotulos e nao numeros
```

```python
pesquisa
```

```python
#saber quantas pessoas avaliaram o playstation como ruim
pesquisa.loc['Playstation4','ruim'] #cruza exatamente o valor da linha com a coluna
```

```python
#acessando valores passando valores em lista dos indices rotulados
pesquisa.loc[['XboxOne', 'Switch']]
```

```python
#retorne todas as linhas e apenas as colunas com rotulos 'bom' e "pessimo"
pesquisa[['bom', 'pessimo']] #para selecionar mais de uma coluna precisa colocar 2 colchetes
```

```python
#retorne todas as linhas e apenas as colunas com rotulos 'bom' e "pessimo" usando metodo loc
pesquisa.loc[:, ['bom', 'pessimo']] #para selecionar mais de uma coluna precisa colocar 2 colchetes usando a função loc
```
