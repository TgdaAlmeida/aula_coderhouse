```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
x = [1,2,3,4,5]
y = [10,8,6,4,2]

plt.plot(x,y)

plt.show()
```

```python
x = ['A','B','C','D','E']
y = [3,7,1,8,4]
plt.bar(x,y)
plt.show()
```

```python
labels = ['Maças', 'Bananas', 'Laranjas', 'Morangos']
quantidades = [20, 25, 10, 30]
plt.pie(quantidades, labels=labels)
plt.show()
```

Grafico de dispersao: mostra a relacao entre duas variaveis numericas.

```python
x = [1, 2, 3, 4, 5]
y = [10, 8, 6, 4, 2]
plt.scatter(x,y)
plt.show()
```

Histograma: mostra a distribuição de frenquencia de dados em um intervalo.

```python
dados = [1,3,3,4,4,4,5,5,6,7,8,9]
plt.hist(dados, bins=4)
plt.show()
```

Area: mostra a tendencia e a magnitude da mudança ao longo do tempo.

```python
plt.plot([1,2,3,4],[5,6,7,8])
plt.title('Titulo do grafico')
plt.show()
```
