```python
import pandas as pd
```

```python
data = pd.read_csv('2004-2021.csv', sep=';')
```

```python
data['MES'] = data['DATA INICIAL'].str.split('-').str[1]
data['ANO'] = data['DATA INICIAL'].str.extract(r'(\d{4})')
```

```python
data['ANO'] = data['ANO'].astype(int)
data['MES'] = data['MES'].astype(int)
```

```python
data_pre = data.copy()
```

```python
data_pre.head(2)
```

**CONVERTENDO AS DATAS** que estão como object

```python
data_pre['DATA INICIAL'] = pd.to_datetime(data_pre['DATA INICIAL'])
data_pre['DATA FINAL'] = pd.to_datetime(data_pre['DATA FINAL'])
```

```python
data_pre.info()
```

**DADOS NUMERICOS**

```python
colunas_a_tratar = ['MARGEM MÉDIA REVENDA', 
                 'PREÇO MÉDIO DISTRIBUIÇÃO', 
                 'DESVIO PADRÃO DISTRIBUIÇÃO', 
                 'PREÇO MÍNIMO DISTRIBUIÇÃO', 
                 'PREÇO MÁXIMO DISTRIBUIÇÃO', 
                 'COEF DE VARIAÇÃO DISTRIBUIÇÃO']
```

```python
for atributo in colunas_a_tratar:
    #converte as colunas de valor string para um tipo numerico
    #em caso de erro na conversao ex. uma string que nao representa numero, um valora NaN sera colocado no lugar
    data_pre[atributo] = pd.to_numeric(data_pre[atributo], errors= "coerce")
```

```python
#apos puxar o info percebe-se que alguns valores passaram a ser NaN, pois na hora da conversao alguns estavam como string em formato que nao da
#pra converter em float64
data_pre.info()
```

**TRATANDO OS DADOS NaN**

```python
#checa cada valor da series se o valor é nulo ou não
mask = data_pre['PREÇO MÉDIO DISTRIBUIÇÃO'].isnull()
mask
```

```python
data_pre[mask]
```

Preenchendo os NaN com um valor constante

```python
#retorna uma copia do DataFrame com todos os valores NaN de todas as colunas agora preenchido com 0
#data alterar o proprio dataframe use o argumento inplace = True
data_pre_fill = data_pre.fillna(0)
data_pre_fill
```

```python
data_pre_fill[mask] #valores que eram "-" agora sao 0
```

**DELETANDO OS VALORES NAN**

```python
#remove no proprio DF todas as linhas com valores NaN em quaisquer colunas
data_pre.dropna(inplace= True)
```
