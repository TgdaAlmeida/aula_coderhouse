```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
fastfood = pd.read_excel('fastfood.xlsx')
```

```python
fastfood2 = fastfood.groupby('restaurant')['sugar'].mean().reset_index()
```

```python
fastfood2
```

```python
fig, ax = plt.subplots()

sns.barplot(x = fastfood2['restaurant'], y = fastfood2['sugar'], data = fastfood2, ax = ax)
ax.set_xticklabels(fastfood2['restaurant'], rotation = 45)
```

```python
sns.histplot(x = fastfood['calories'], data = fastfood)
```

```python
#x = np.random.normal(size = 100)
#fig, ax = plt.subplots()
sns.displot(x = fastfood2['sugar'], kde = True) #kde cria uma curva ao longo do diagrama
plt.show()
```
