```python
data['ESTADO'].unique() #tira as repetiçoes e deixa somente os valores unicos
```

```python
# faz uma comparação elemente a elemento da series, retornando uma series de booleans
data['ESTADO'] == 'SAO PAULO'
```

```python
#salvando a series de booleans em uma variavel
selecao = data['ESTADO'] == 'SAO PAULO'
```

```python
#retorna somente os valores da linha da variavel cujo valor seja TRUE
postos_sp = data[selecao]
```

USANDO LOC

```python
#faz a mesma coisa que a de cima, so retorna os valores TRUE da variavel
data.loc[selecao]
```

USANDO O METODO QUERY

```python
#filtra as linhas onde o valor é igual a "SAO PAULO"
data.query('ESTADO == "SAO PAULO"') #pra usar string dentro de outra é preciso usar aspas diferentes
```

```python
postos_sp.head(2)
```

```python
#o python cria uma coluna com os indeces antigos
postos_sp.reset_index(drop=True, inplace= True) #se usar o drop=True dentro dos parenteses o python nao cria essa coluna e o implace true 
#ja faz a modificação direto no dataframe
```
